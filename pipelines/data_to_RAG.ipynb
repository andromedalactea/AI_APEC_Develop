Import Required Libraries

```python
# Python Libraries
import os
from collections import Counter, defaultdict
from IPython.display import display, clear_output
import time

# External Libraries
import numpy as np
```

Initial Analize of the data

```python

# Ruta base de la unidad montada en WSL
base_path = '/mnt/i'

# Inicializamos un contador para las extensiones y un diccionario para tama침os
file_types = Counter()
file_sizes = defaultdict(int)  # Usaremos un defaultdict para acumular los tama침os

# Funci칩n para mostrar resultados parciales
def mostrar_resultados_parciales(file_types, file_sizes):
    clear_output(wait=True)  # Limpiamos la salida para actualizar
    print("Conteo de archivos y tama침o por extensi칩n (actualizado):")
    for ext, count in file_types.items():
        size_in_mb = file_sizes[ext] / (1024 * 1024)  # Convertir tama침o a MB
        print(f"{ext}: {count} archivos, {size_in_mb:.2f} MB")

# Recorremos todas las carpetas y subcarpetas usando os.walk
for dirpath, dirnames, filenames in os.walk(base_path):
    for filename in filenames:
        # Extraer la extensi칩n del archivo
        file_extension = os.path.splitext(filename)[1].lower()
        # Obtener la ruta completa del archivo
        file_path = os.path.join(dirpath, filename)
        
        try:
            # Obtener el tama침o del archivo
            file_size = os.path.getsize(file_path)
            # Contar la extensi칩n y acumular el tama침o
            file_types[file_extension] += 1
            file_sizes[file_extension] += file_size
        except FileNotFoundError:
            # Si el archivo no est치 accesible o es eliminado mientras se recorre, ignorarlo
            continue
        
    # Mostrar resultados parciales despu칠s de cada directorio
    mostrar_resultados_parciales(file_types, file_sizes)
    time.sleep(0.1)  # A침adir una peque침a pausa para que la actualizaci칩n sea m치s visible

# Mostrar resultados finales
print("Conteo final de archivos y tama침o por extensi칩n:")
for ext, count in file_types.items():
    size_in_mb = file_sizes[ext] / (1024 * 1024)  # Convertir tama침o a MB
    print(f"{ext}: {count} archivos, {size_in_mb:.2f} MB")
```
